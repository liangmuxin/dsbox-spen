```python
def _generate_output(labels, res):
    output = []
    for v in range(len(res)):
        tmp = []
        for i, r in enumerate(list(res[v,:])):
            if r == 1.0:
                tmp.append(str(labels[i]))
        val = ",".join(tmp)
        output.append(val)
    return output
```
